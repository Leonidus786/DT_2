# Introduction to Statistics - Detailed Notebook
This notebook covers all major statistical concepts with explanations, formulas, and plots.

## 1. Sample vs Population
Population = full data, Sample = subset used for analysis.

```python
import numpy as np

# Create population
population = np.arange(1000)

# Take a sample
sample = np.random.choice(population, size=100, replace=False)

print("Population size:", len(population))
print("Sample size:", len(sample))
```

## 2. Arithmetic Mean
Formula: mean = sum(x)/n

```python
data = [10, 20, 30, 40]
mean = np.mean(data)
print("Mean:", mean)
```

## 3. Standard Deviation
Measures spread of data

```python
import matplotlib.pyplot as plt

# using a normal distributed data
data = np.random.normal(50, 10, 1000)

mean = np.mean(data)
std = np.std(data)

print("Mean:", mean)
print("Std Dev:", std)

plt.hist(data)
plt.title("Histogram showing spread")
plt.show()
```

## 4. Normal Distribution

```python
from scipy.stats import norm

# basically you are generating 100 evenly spaced numbers starting form -3 and ending at 3
# for eg. -3.0 , -2.987, 0 , 0.5453, 1
x = np.linspace(-3, 3, 100)
# pdf == probability density functions
y = norm.pdf(x)

plt.plot(x, y)
plt.title("Normal Distribution Curve")
plt.show()
```

## 5. Uniform Distribution

```python
from scipy.stats import uniform

data = uniform.rvs(size=1000)

plt.hist(data)
plt.title("Uniform Distribution")
plt.show()
```

## 6. Skewed Distribution

```python
import seaborn as sns
```

```python
data = np.random.exponential(scale=1, size=1000)

sns.histplot(data,kde=True)
plt.title("Right Skewed Distribution")
plt.show()
```

## 8. T-Test

```python
from scipy.stats import ttest_1samp

data = [45, 50, 55, 60]
t_stat, p = ttest_1samp(data, 50)

print("p-value:", p)
```

## 9. ANOVA

```python
from scipy.stats import f_oneway

g1 = [10,20,30]
g2 = [15,25,35]
g3 = [20,30,40]

f_stat, p = f_oneway(g1,g2,g3)
print("p-value:", p)
```

## 10. Correlation

```python
x = [1,2,3,4]
y = [2,4,6,8]

corr = np.corrcoef(x,y)
print(corr)
```
