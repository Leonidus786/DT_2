## Arithmetic Mean Derivation
Mean = sum of observations / number of observations.
Derived from balancing point of data.

```python
import numpy as np

data = [12, 15, 18, 20, 25]

mean = sum(data)/len(data)
print("Manual Mean:", mean)
print("Numpy Mean:", np.mean(data))
```

## Chi-Square Test Example
Testing independence between gender and product preference

```python
from scipy.stats import chi2_contingency

table = [[30, 20],
         [20, 30]]

chi2, p, dof, expected = chi2_contingency(table)

print("Chi2:", chi2)
print("p-value:", p)
print("Expected:", expected)
```

## Z-Test Manual Calculation

```python
import numpy as np

data = [50, 52, 48, 47, 49, 51]
mu = 50
sigma = 2

x_bar = np.mean(data)
n = len(data)

z = (x_bar - mu) / (sigma / np.sqrt(n))

print("Z-statistic:", z)
```

## Central Limit Theorem Simulation

```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2, size=10000)

sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means)
plt.title("CLT: Distribution of Sample Means")
plt.show()
```
